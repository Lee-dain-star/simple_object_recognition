```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img11 = cv2.imread('11.jpg')
img12 = cv2.imread('12.jpg')
img13 = cv2.imread('13.jpg')

plt.figure(figsize=(16,8))
plt.subplot(1,3,1)
plt.imshow(img11, cmap='gray'), plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(img12, cmap='gray'), plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(img13, cmap='gray'), plt.axis('off')
```

```python
kernel = np.ones((30,30), np.uint8)
objects = cv2.morphologyEx(img11, cv2.MORPH_OPEN, kernel)
kernel = np.ones((10, 10), np.uint8)
objects = cv2.morphologyEx(objects, cv2.MORPH_CLOSE, kernel)

gray = cv2.cvtColor(objects, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize=(16,8))
plt.imshow(thresh, cmap='gray'), plt.axis('off')
```

```python
a = [] #Area
p = [] #perimeter
f = [] #Formfactor
```

```python
contours = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours[1][:-1]:
    
    img1 = cv2.drawContours(img11, [cnt], 0, (0,255,0), 3)
    
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt,True)
    form = (4*np.pi*area)/(perimeter**2)
    
    a.append(area)
    p.append(perimeter)
    f.append(form)
    
    text = f"area : {int(area)}"
    text2 = f"peri : {int(perimeter)}"
    text3 = f"form : {round(form,3)}"
    
    print(text)
    print(text2)
    print(text3)
    print('-----------------')

plt.figure(figsize=(16,8))    
plt.imshow(img1)
```

```python
c1 = cv2.imread('c1.jpg')
c2 = cv2.imread('c2.jpg')
c3 = cv2.imread('c3.jpg')
```

```python
img = [c1,c2,c3]

for i in range(len(img)):
    
    kernel = np.ones((50,50), np.uint8)
    objects = cv2.morphologyEx(img[i], cv2.MORPH_OPEN, kernel)
    kernel = np.ones((10, 10), np.uint8)
    objects = cv2.morphologyEx(objects, cv2.MORPH_CLOSE, kernel)

    gray = cv2.cvtColor(objects, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    contours = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contours[1][:-1]:

        img[i] = cv2.drawContours(np.array(img[i]), [cnt], 0, (0,255,0), 3)

        area = cv2.contourArea(cnt)
        perimeter = cv2.arcLength(cnt,True)   
        form = (4*np.pi*area)/(perimeter**2)
        
        if int(area) == 139:
            continue
        
        a.append(area)
        p.append(perimeter)
        f.append(form)
        
        text = f"area : {int(area)}"
        text2 = f"peri : {int(perimeter)}"
        text3 = f"form : {round(form,3)}"
        
        print(text)
        print(text2)
        print(text3)
        print('-----------------')

    plt.figure(figsize=(12,6))  
    plt.subplot(2,5,i+1)
    plt.imshow(img[i]), plt.axis('off')
```

```python
import pandas as pd

#label = ['pencil','circle','fork','eraser','lipstick','shading','tube','mouse','stick','stamp case']
df= pd.DataFrame({'Area': a, 'Form':f})
df.loc[(df['Area']>250000) & (df['Form']>0.7), 'object'] = 'mouse'
df.loc[(200000>df['Area'])& (df['Area']>150000) & (df['Form']>0.8), 'object'] = 'shading'
df.loc[(150000>df['Area'])&(df['Area']>100000) & (df['Form']>0.4), 'object'] = 'stamp case'
df.loc[(100000>df['Area'])& (df['Area']>50000) & (df['Form']>0.6), 'object'] = 'eraser'
df.loc[(100000>df['Area'])&(df['Area']>60000) & (df['Form']>0.49) & (df['Form']<0.6), 'object'] = 'lipstick'
df.loc[(100000>df['Area'])&(df['Area']>62000) & (df['Form']>0.4) & (df['Form']<0.49), 'object'] = 'tube'
df.loc[(100000>df['Area'])&(df['Area']>63000) & (df['Form']>0.2) & (df['Form']<0.3), 'object'] = 'pen'
df.loc[(62000>df['Area'])&(df['Area']>55000) & (df['Form']>0.85) & (df['Form']<0.95), 'object'] = 'circle'
df.loc[(63000>df['Area'])&(df['Area']>59000) & (df['Form']<0.22) & (df['Form']>0.1), 'object'] = 'stick'
df.loc[(60000>df['Area'])&(df['Area']>50000) & (df['Form']<0.2) & (df['Form']>0.05), 'object'] = 'fork'
df.loc[(df['Area']<50000) & (df['Form']<0.2), 'object'] = 'pencil'
df = df[df['object']!='pen']
df
```

```python
import math
from scipy.spatial import distance

def Object_Recognition(avg, img):    
    kernel = np.ones((20,20), np.uint8)
    objects = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    kernel = np.ones((10, 10), np.uint8)
    objects = cv2.morphologyEx(objects, cv2.MORPH_CLOSE, kernel)
    
    gray = cv2.cvtColor(objects, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    
    print(plt.imshow(thresh,'gray'))
    
    contours = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    
    for cnt in contours[1][:-1]:
        
        img = cv2.drawContours(img, [cnt], 0, (0,255,0), 3)
        
        M = cv2.moments(cnt)
        center = (int(M["m10"] / M["m00"]),int(M["m01"] / M["m00"]))

        area = cv2.contourArea(cnt)
        
        if int(area) < 20000:
            continue
            
        perimeter = cv2.arcLength(cnt,True)   
        form = (4*np.pi*area)/(perimeter**2) 
            
        c=[]
        label = ['circle','eraser','fork','lipstick','mouse','pencil','shading','stamp case','stick','tube']
        x = [avg.Area, avg.Form]

        for i in range(10):
            c.append(distance.euclidean((x[0][i],x[1][i]),(area/270000,form)))
#             c.append(math.sqrt((area-x[0][i])**2+(form-x[1][i])**2))
        
        print(label[np.argmin(c,0)])
        text = f'{label[np.argmin(c,0)]}'
        img = cv2.putText(img, text, center, cv2.FONT_HERSHEY_SIMPLEX, 3, (0,127,255),3)
        
    plt.figure(figsize=(12,6))  
    plt.imshow(img), plt.axis('off')
```

```python
def capture():
    cap = cv2.VideoCapture(0)
    cnt=0
    no=1
    while True:
        ret, frame = cap.read()

        if not ret:
            continue

        cnt+=1
        if(cnt==10):
            cv2.imwrite(f'./image/test{no}.jpg',frame)
            print(f'save{no}')
            no+=1
            break

        cv2.imshow('Frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
```

```python
def video():
    cap = cv2.VideoCapture(0)
    
    print(cap.get(3), cap.get(4))

    while True:
        ret, frame = cap.read()

        if not ret:
            continue
        
        frame = Object_Recognition(frame)
        cv2.imshow('Frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()
```
